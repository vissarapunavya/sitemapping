```python
train_imgs = len(os.listdir("/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/train"))
train_labels = len(os.listdir("/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/train"))
valid_imgs = len(os.listdir("/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/valid"))
valid_labels = len(os.listdir("/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/valid"))

print("Train images:", train_imgs, "Train labels:", train_labels)
print("Valid images:", valid_imgs, "Valid labels:", valid_labels)
```

```python
yaml_content = """
train: /content/drive/MyDrive/vegetation segmentation.v2i.yolov8/train
val: /content/drive/MyDrive/vegetation segmentation.v2i.yolov8/valid

nc: 3
names: ['vegetation', 'structure', 'erosion']
"""

yaml_path ="/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/data.yaml"
with open(yaml_path, "w") as f:
    f.write(yaml_content)

print("data.yaml created at:", yaml_path)
```

```python
#from scratch
model_scratch= YOLO("yolov8x.yaml")
model_scratch.train(
    data=yaml_path,
    epochs=50,
    imgsz=640,
    batch=16,
    name="archaeo_yolov8_scratch"
)
```

```python
#  pre-trained YOLOv8 model
model = YOLO("yolov8x.pt")
model.train(
    data=yaml_path,
    epochs=50,
    imgsz=640,
    batch=16,
    name="archaeo_yolov8"
)
```

```python

print("Pretrained metrics:")
pretrained_metrics = model.val()
print(pretrained_metrics)



```

```python
print("Scratch metrics:")
scratch_metrics = model_scratch.val()
print(scratch_metrics)
```

```python
import matplotlib.pyplot as plt
import cv2
import os
import random
model_scratch= YOLO("yolov8x.yaml")

# Pick 2 random images from your test directory for visualization
test_dir = f"/content/drive/MyDrive/vegetation segmentation.v2i.yolov8/test/images"
test_images = os.listdir(test_dir)
sample_images = random.sample(test_images, 2)

# Helper to load and plot images
def plot_predictions(image_path, title):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis('off')
    plt.title(title)

for fname in sample_images:
    test_img_path = os.path.join(test_dir, fname)

    # Pretrained model inference
    pre_out = model.predict(source=test_img_path, save=False, conf=0.25)
    pre_img = pre_out[0].plot()  # Visualize with bboxes using Ultralytics plotting

    # Scratch model inference
    scratch_out = model_scratch.predict(source=test_img_path, save=False, conf=0.25)
    scratch_img = scratch_out[0].plot()

    # Plot side by side
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.imshow(pre_img)
    plt.title("Pretrained YOLOv8")
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.imshow(scratch_img)
    plt.title("Scratch-trained YOLOv8")
    plt.axis('off')

    plt.suptitle(f"Predictions for {fname}")
    plt.show()
```
